# Analysis of players at different positions in fifa19

```python
import pandas as pd 
import numpy as np
data = pd.read_csv('./data/data.csv')
print(len(data))
print(data.shape)
```

### Top players at attack

```python
forward_positions = ['ST','RF','CF','LF','LS','RS']
forward_player_list = []
forward_overall = []
fwd_pos = []
for each in forward_positions:
    forward_pos_bool = data.loc[:,'Position'] == each
    forward_pos_df = data.loc[forward_pos_bool,'Name']
    forward_pos_list = list(forward_pos_df)
    forward_overall_df = data.loc[forward_pos_bool,'Overall']
    forward_overall_list = list(forward_overall_df)
    fwd_pos_df = data.loc[forward_pos_bool,'Position']
    fwd_pos_list = list(fwd_pos_df)
    for player in forward_pos_list:
        forward_player_list.append(player)
    for num in forward_overall_list:
        forward_overall.append(num)
    for pos in fwd_pos_list:
        fwd_pos.append(pos)
forward_player_df = pd.DataFrame({'Players':forward_player_list})
forward_player_df['Overall'] = forward_overall
forward_player_df['Position'] = fwd_pos
num_fwd = 30
print(' The Top ' + str(num_fwd) + ' attackers are')
print(forward_player_df.sort_values(by=['Overall'],ascending=False).head(num_fwd))
```

### Top players at Defence

```python
defence_positions = ['CB','LB','RB','RCB','LCB','RWB','LWB']
defence_player_list = []
defence_overall = []
def_pos = []
for each in defence_positions:
    defence_pos_bool = data.loc[:,'Position'] == each
    defence_pos_df = data.loc[defence_pos_bool,'Name']
    defence_pos_list = list(defence_pos_df)
    defence_overall_df = data.loc[defence_pos_bool,'Overall']
    defence_overall_list = list(defence_overall_df)
    def_pos_df = data.loc[defence_pos_bool,'Position']
    def_pos_list = list(def_pos_df)
    for player in defence_pos_list:
        defence_player_list.append(player)
    for num in defence_overall_list:
        defence_overall.append(num)
    for pos in def_pos_list:
        def_pos.append(pos)        
defence_player_df = pd.DataFrame({'Players':defence_player_list})
defence_player_df['Overall'] = defence_overall
defence_player_df['Position'] = def_pos
num_def = 30
print(' The Top ' + str(num_def) + ' defenders are')
print(defence_player_df.sort_values(by=['Overall'],ascending=False).head(num_def))
    
```

### The Top players at Mid 

```python
midfield_positions = ['ST','RF','CF','LF','LS','RS']
midfield_player_list = []
midfield_overall = []
mid_pos = []
for each in midfield_positions:
    midfield_pos_bool = data.loc[:,'Position'] == each
    midfield_pos_df = data.loc[midfield_pos_bool,'Name']
    midfield_pos_list = list(midfield_pos_df)
    midfield_overall_df = data.loc[midfield_pos_bool,'Overall']
    midfield_overall_list = list(midfield_overall_df)
    mid_pos_df = data.loc[midfield_pos_bool,'Position']
    mid_pos_list = list(mid_pos_df)
    for player in midfield_pos_list:
        midfield_player_list.append(player)
    for num in midfield_overall_list:
        midfield_overall.append(num)
    for pos in mid_pos_list:
        mid_pos.append(pos)
midfield_player_df = pd.DataFrame({'Players':midfield_player_list})
midfield_player_df['Overall'] = midfield_overall
midfield_player_df['Position'] = mid_pos
num_mid = 30
print(' The Top ' + str(num_mid) + ' midfielders are')
print(midfield_player_df.sort_values(by=['Overall'],ascending=False).head(num_mid))
```

### The Top Goalkeepers 

```python
goalkeeper_player_list=[]
goalkeeper_overall = []
goalkeeper_pos_bool = data.loc[:,'Position'] == 'GK'
goalkeeper_pos_df = data.loc[goalkeeper_pos_bool,'Name']
goalkeeper_pos_list = list(goalkeeper_pos_df)
goalkeeper_overall_df = data.loc[goalkeeper_pos_bool,'Overall']
goalkeeper_overall_list = list(goalkeeper_overall_df)
for player in goalkeeper_pos_list:
    goalkeeper_player_list.append(player)
for num in goalkeeper_overall_list:
    goalkeeper_overall.append(num)
goalkeeper_player_df = pd.DataFrame({'Players':goalkeeper_player_list})
goalkeeper_player_df['Overall'] = goalkeeper_overall
num_gk = 20
print('The top '+ str(num_gk)+ ' Goalkeepers are :')
print(goalkeeper_player_df.sort_values(by=['Overall'],ascending=False).head(num_gk))
```

### Top 5 Players at specific positions

```python
player_pos_list = []
position_df = data.loc[:,'Position'].value_counts().index
for each in position_df:
    position_bool = data.loc[:,'Position'] == each
    players_position = data.loc[position_bool,'Name']
    player_pos_list.append(players_position)
position_list = position_df.tolist()
dictionary=dict(zip(position_list,player_pos_list))
num_player = 5
for each in position_df:
    print('Top 5 players at Position '+ each+ ' are:')
    for num in range(0,num_player):
        print(dictionary[each].values[num])
```
